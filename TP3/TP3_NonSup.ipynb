```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np

data = pd.read_csv('./villes.csv', sep=';')
X = data.iloc[:, 1:13].values  
labels = data.iloc[:, 0].values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Calculate the number of components to retain for 70% of information
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance_ratio = explained_variance_ratio.cumsum()
n_components_to_retain = len(cumulative_variance_ratio[cumulative_variance_ratio <= 0.70])

# Interpretation of the first two principal axes
first_axis_explained_variance = explained_variance_ratio[0]
second_axis_explained_variance = explained_variance_ratio[1]

print(f"Number of components to retain for 70% of information: {n_components_to_retain}")
print(f"Explained variance of the first axis: {first_axis_explained_variance:.2f}")
print(f"Explained variance of the second axis: {second_axis_explained_variance:.2f}")

plt.scatter(X_pca[:, 0], X_pca[:, 1])

for label, x, y in zip(labels, X_pca[:, 0], X_pca[:, 1]):
    plt.annotate(label, xy=(x, y), xytext=(-0.2, 0.2), textcoords='offset points')

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of French Cities')
plt.show()
```

```python
# Calculate the explained variance for each principal component
explained_variance = pca.explained_variance_ratio_

# Create a bar graph to show the explained variance for each component
plt.bar(range(1, len(explained_variance) + 1), explained_variance)
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance')
plt.title('Explained Variance for Each Principal Component')
plt.xticks(range(1, len(explained_variance) + 1))
plt.show()
```

**Analysis**
Only two componants are necessary to accumulate avariance above 90% !! impressive

```python
# Get the principal component loadings for each feature
principal_component_loadings = pca.components_

# Create a bar graph to show feature contributions for the first two principal components
n_principal_components = 2  # You can change this to visualize more components if needed
n_features = len(principal_component_loadings)

# Plot the feature contributions for each of the first two principal components
for i in range(n_principal_components):
    plt.figure(figsize=(10, 6))
    plt.bar(range(n_features), principal_component_loadings[i])
    plt.xlabel('Feature (Month)')
    plt.ylabel('Feature Contribution')
    plt.title(f'Feature Contributions for Principal Component {i+1}')
    plt.xticks(range(n_features), range(1, n_features+1))
    plt.show()
```

**Analysis**
January and December represent the two dominating features for the seconf componant
the first compannt has an over all stable contribution by all the features

Steps for the study:
Data Import and Preparation:

The code starts by importing the necessary libraries: pandas for data handling, matplotlib for plotting, and two modules from scikit-learn: PCA for Principal Component Analysis and StandardScaler for data preprocessing.

The temperature data is extracted into the X variable, and the city names are extracted into the labels variable.

Principal Component Analysis (PCA):

Principal Component Analysis is a dimensionality reduction technique that aims to capture the most important information in a dataset by projecting it onto a new set of orthogonal axes (principal components).

StandardScaler is used to standardize the data, which means it scales the features to have zero mean and unit variance.

PCA is performed on the standardized data using PCA(), and the result is stored in X_pca.

The code then calculates the number of principal components required to retain at least 70% of the information. It does this by computing the explained variance ratio for each component and accumulating it until it reaches 70%.

It also provides an interpretation of the first two principal axes by printing the explained variance for each axis.

Visualization of PCA Results:

The code creates a scatter plot to visualize the cities projected onto the first two principal components.

It iterates through the city names and their corresponding coordinates in the PCA space and annotates the data points with the city names.

The x-axis represents the first principal component, and the y-axis represents the second principal component.

Finally, it adds labels, a title, and displays the plot using plt.show().


```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

data = pd.read_csv('./crimes.csv', sep=';')
X = data.iloc[:, 1:13].values  
labels = data.iloc[:, 0].values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Calculate the number of components to retain for 70% of information
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance_ratio = explained_variance_ratio.cumsum()
n_components_to_retain = len(cumulative_variance_ratio[cumulative_variance_ratio <= 0.70])

# Interpretation of the first two principal axes
first_axis_explained_variance = explained_variance_ratio[0]
second_axis_explained_variance = explained_variance_ratio[1]

print(f"Number of components to retain for 70% of information: {n_components_to_retain}")
print(f"Explained variance of the first axis: {first_axis_explained_variance:.2f}")
print(f"Explained variance of the second axis: {second_axis_explained_variance:.2f}")

plt.scatter(X_pca[:, 0], X_pca[:, 1])

for label, x, y in zip(labels, X_pca[:, 0], X_pca[:, 1]):
    plt.annotate(label, xy=(x, y), xytext=(-0.2, 0.2), textcoords='offset points')

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('crimes in cities from the United States')
plt.show()
```

```python


# Calculate the explained variance for each principal component
explained_variance = pca.explained_variance_ratio_

# Create a bar graph to show the explained variance for each component
plt.bar(range(1, len(explained_variance) + 1), explained_variance)
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance')
plt.title('Explained Variance for Each Principal Component')
plt.xticks(range(1, len(explained_variance) + 1))
plt.show()

# Get the principal component loadings for each feature
principal_component_loadings = pca.components_

# Create a bar graph to show feature contributions for the first two principal components
n_principal_components = 2  # You can change this to visualize more components if needed
n_features = len(principal_component_loadings)

# Plot the feature contributions for each of the first two principal components
for i in range(n_principal_components):
    plt.figure(figsize=(10, 6))
    plt.bar(range(n_features), principal_component_loadings[i])
    plt.xlabel('Feature')
    plt.ylabel('Feature Contribution')
    plt.title(f'Feature Contributions for Principal Component {i+1}')
    plt.xticks(range(n_features), range(1, n_features+1))
    plt.show()
```
